```python
mybam = pysam.Samfile('data/MqsR2h_cutting/MqsR2h_PNK_sorted.bam', "rb")
pairedreads = pysam.Samfile("gi|556503834|ref|NC_000913.3|", "wb", template=mybam)
for read in mybam.fetch():
     if read.is_paired:
             pairedreads.write(read)

pairedreads.close()
mybam.close()
```

```python
mybam = pysam.Samfile('data/MqsR2h_cutting/MqsR2h_PNK_sorted.bam' ,"rb")
for pileupcolumn in mybam.pileup( 'gi|556503834|ref|NC_000913.3|', 100, 120):
    print
    print 'coverage at base %s = %s' % (pileupcolumn.pos , pileupcolumn.n)
    for pileupread in pileupcolumn.pileups:
    
        print '\tbase in read %s = %s' % (pileupread.alignment.qname, pileupread.alignment.seq[pileupread.qpos])

mybam.close()
```
